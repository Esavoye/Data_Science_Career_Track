```python
# adding length of comments and word count of each comment
unpickled_df['comment_len'] = unpickled_df['clean_text'].astype(str).swifter.apply(len)
unpickled_df['word_count'] = unpickled_df['clean_text'].swifter.apply(lambda x: len(str(x).split()))
unpickled_df['word_split'] = unpickled_df['clean_text'].swifter.apply(lambda x: str(x).split())
```

```python
#Check stopwords currently in list
print(stopword_list)
```

```python
# Create function to strip punctuation
def strip_punctuation(s):
    return ''.join(c for c in s if c not in punctuation)
```

```python
#update stopword list to remove punctuation
stopword_list_updated = [strip_punctuation(word) for word in stopword_list] 
```

```python
#Check stopwords currently in list
print(stopword_list_updated)
```

```python
# All reviews:
text = " ".join(comment for comment in unpickled_df.clean_text)
print ("There are {} words in the combination of all comments.".format(len(text)))

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Using a word cloud doesn't really give much insight. Looking at the image above, I believe I can remove the stop words listed previously to clean up possible noise.

```python
# make list of all the words across all comments
all_words_unp = list(itertools.chain(*unpickled_df['word_split']))
```

```python
# Create counter
counts_all_words_unp = collections.Counter(all_words_unp)

counts_all_words_unp.most_common(15)
```

Looking at the most common words, they all appear to be noise created by stop words. By removing them, the noise should decrease.

```python
# function to remove stop words
def remove_stopwords(list_of_words):
    words = [w for w in list_of_words if not w in stopword_list_updated]
    return words
```

```python
unpickled_df['clean_text_no_stpwd'] = unpickled_df['word_split'].swifter.apply(remove_stopwords)
```

```python
unpickled_df.clean_text.head()
```

We'll take another look at the 15 most frequent words and the word cloud without the any of the stopwords.

```python
# All reviews:
text = " ".join(word for comment in unpickled_df.clean_text_no_stpwd for word in comment)
print ("There are {} words in the combination of all comments.".format(len(text)))

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
